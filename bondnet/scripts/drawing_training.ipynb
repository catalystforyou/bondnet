```python
from matplotlib import pyplot as plt
```

```python
training_log = open('train.log').read().splitlines()[1:]
```

```python
train_acc = [float(log.split()[2]) for log in training_log]
val_acc = [float(log.split()[3]) for log in training_log]
```

```python
plt.plot(train_acc, label='Train')
plt.plot(val_acc, label='Val')
plt.legend()
plt.xlabel('epoch')
plt.ylabel('MAE/eV')
```
